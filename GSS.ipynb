```python
# Homework 1, MATH/CSCI 147, due Friday 8/12/24
# Billy Zhang

# TODO: Tinker with values for these global variables as you experiment!

import random
import time
import math

a0 = 0
d0 = 5000 # size of initial search domain
count = 0 # test code should set this to zero before every run of any algorithm
opt = 0 # test code should randomize this before each run
x_tol = 0.001 # stopping condition for all algorithms
epsilon = x_tol * 0.001 # for binary_search
beta = 1.5-math.sqrt(5)/2 # for golden_section_search

eval_time = .01 # amount of time it takes to evaluate f
other_time = .001 # amount of time it takes to access the lookup table


def f(x):
    time.sleep(eval_time) # delay
    global count
    count += 1
    if x-opt>0: return x-opt
    return opt-x

tableDigits = 3 # how many digits can be rounded
tableHit = 0
tableTotal = 0

def get_f_eval(x,f_table):
    global tableHit,tableTotal
    tableTotal +=1
    time.sleep(other_time) # delay
    if x in f_table:
        tableHit+=1
        return f_table[x]
    fx = f(x)
    f_table[x] = fx
    return fx

def get_f_eval_opt(x,f_table):
    global tableHit,tableTotal
    x = math.floor(x * (10**tableDigits))/(10**tableDigits)
    tableTotal +=1
    time.sleep(other_time) # delay
    if x in f_table:
        tableHit+=1
        return f_table[x]
    fx = f(x)
    f_table[x] = fx
    return fx

def reset():
    # count, time, ftable, table hits, table total
    return 0, time.time(),
def resetgold():
    # count, time, ftable, table hits, table total
    return 0, time.time(), {}, 0,0
```

```python
# Implement these three algorithms

def ternary_search(a,d):
    interval = (d-a)/3
    if(interval < x_tol): return a
    b = a + interval
    c = d - interval
    if f(b) < f(c):
      return ternary_search(a,c)
    return ternary_search(b,d)

def binary_search(a,d):
    interval = (d-a)/2
    if(interval < x_tol): return a
    b = a + interval
    c = a + interval + epsilon
    if f(b) < f(c):
      return binary_search(a,c)
    return binary_search(b,d)

def golden_section_search(a,d,f_table,rounding=False):
    interval = (d-a) * beta
    if(interval < x_tol): return a
    b = a + interval
    c = d - interval
    fb = get_f_eval_opt(b,f_table) if rounding else get_f_eval(b,f_table)
    fc = get_f_eval_opt(c,f_table) if rounding else get_f_eval(c,f_table)

    if fb < fc:
      return golden_section_search(a,c,f_table,rounding)
    return golden_section_search(b,d,f_table,rounding)
```
